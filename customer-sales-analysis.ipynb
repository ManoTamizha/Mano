```python
df['Review Rating'] = df.groupby('Category')['Review Rating']\
                 .transform(lambda x: x.fillna(x.mean()))
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
print(df.columns)
```

```python
df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'}, inplace=True)
```

```python
df['purchase_amount'] = pd.to_numeric(df['purchase_amount'], errors='coerce')
```

```python
df.drop_duplicates(inplace=True)
```

```python
df = df[df['purchase_amount'] < df['purchase_amount'].quantile(0.99)]
```

```python
df['subscription_status'] = df['subscription_status'].map({'yes':1, 'no':0})
df['discount_applied'] = df['discount_applied'].map({'yes':1, 'no':0})
df['promo_code_used'] = df['promo_code_used'].map({'yes':1, 'no':0})
```

```python
df = df[(df['age'] > 10) & (df['age'] < 100)]
```

```python
df['frequency_of_purchases'] = df['frequency_of_purchases'].str.lower()
```
